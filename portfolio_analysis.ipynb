```python
import yfinance as yf
import matplotlib.pyplot as plt
#%matplotlib widget
import pprint
import pandas as pd
```

```python

ticker_name = "AAPL"
start_date = "2024-01-01"
end_date = "2025-01-01"
period_range = "2y" # Valid periods: 1d,5d,1mo,3mo,6mo,1y,2y,5y,10y,ytd

# Download historical data
data = yf.download(ticker_name, period=period_range, auto_adjust=True)
print(data)

# Ticker information
ticker_obj = yf.Ticker(ticker_name)
ticker_info = ticker_obj.info
#pprint.pprint(ticker_info)



# Plotting the closing prices
if 'Close' in data.columns:
    data['Close'].plot(title=f"{ticker_name} Closing Prices")
    plt.xlabel("Date")
    plt.ylabel("Price (USD)")
    plt.grid(True)
    plt.show()
else:
    print("No 'Close' column in the data to plot.")
```
